```python
import pandas as pd
import matplotlib.pyplot as plt
  
# Create a list of data
# to be represented in x-axis
days = [ 'Monday' , 'Tuesday' , 'Wednesday ',  'Thursday' , 'Friday' , 'Saturday' ,'Sunday']
  
# Create a list of data
# to be represented in y-axis
casual = [ 194972, 202326 , 218113 , 233764 , 258174 , 350968 , 289551 ]
  
# Create second list of data to be represented in y-axis
members = [ 374370, 438804 , 456845 , 442114 , 387063 , 343717 , 303828]
  
# Create a dataframe using the two lists
df_days_calories = pd.DataFrame(
    { 'days' : days , 
     'casual': casual , 
     'members': members})
  
ax = plt.gca()
  
#use plot() method on the dataframe
df_days_calories.plot( x = 'days' , y = 'casual', ax = ax )
df_days_calories.plot( x = 'days' , y = 'members' , ax = ax )
```

```python
import pandas as pd
  
# DataFrame of each student and the votes they get
dataframe = pd.DataFrame({'Riders': ['Casual', 'Members'],
                          'amount_of_riders': [1747868,2746741]})
  
# Defining colors for the pie chart
colors = ['pink','steelblue']
  
# Define the ratio of gap of each fragment in a tuple
explode = (0.05, 0.05)
  
# Plotting the pie chart for above dataframe
dataframe.groupby(['Riders']).sum().plot(
    kind='pie', y='amount_of_riders', autopct='%1.0f%%',
  colors=colors, explode=explode)
```
