```python
#After import all 12 CSV files, I concat all of them into one file and named it df 13
df13= pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12])


df13
```

```python
#Dropping empty values from df 13
df13=df13.dropna()
df13
```

```python
#Dropping unecessary columns for analysis. I also dropped duration because some data in 'duration' (excel version) shows as
# #### instead of showing time. So I decided to drop it and re-do it here in Pandas later. I also make a new dataframe 
#for the new file without the dropped columns, df14.
df14= df13.drop(['start_lat','start_lng', 'end_lat','end_lng', 'duration'], axis=1)

df14
```

```python
#Now, I am going to convert 'started_at' and 'ended_at' to datetime. I also add a new column 'duration' and substract
#'started_at' from 'ended_at'

import pandas as pd
import datetime

df14['started_at'] = pd.to_datetime(df14['started_at'], format="%m/%d/%y %H:%M")

df14['ended_at'] = pd.to_datetime(df14['ended_at'], format="%m/%d/%y %H:%M")

df14['duration']= df14['ended_at'] - df14['started_at']

#Making a new dataframe to filter the casual and members. 

df_member= df14[df14['member_casual'].str.contains('member')]
df_casual= df14[df14['member_casual'].str.contains('casual')]

```

```python
#Using the new dataframe to find the mean, max, min of each time duration. 

df_member['duration'].mean() #0 days, 12 minutes  = 12 minutes
df_member['duration'].max() #1 day, 58 minutes = 1498 minutes
df_member['duration'].min() #0 minutes = 0 minutes

df_casual['duration'].mean() #0 days, 22 minutes  = 22 minutes
df_casual['duration'].max() #22 days, 5 hours 55 minutes = 32,035 minutes
df_casual['duration'].min() #0 minutes = 0 minutes
```

```python
import pandas as pd
import matplotlib.pyplot as plt
  
# Create a list of data
# to be represented in x-axis
days = [ 'Monday' , 'Tuesday' , 'Wednesday ',  'Thursday' , 'Friday' , 'Saturday' ,'Sunday']
  
# Create a list of data
# to be represented in y-axis
casual = [ 194972, 202326 , 218113 , 233764 , 258174 , 350968 , 289551 ]
  
# Create second list of data to be represented in y-axis
members = [ 374370, 438804 , 456845 , 442114 , 387063 , 343717 , 303828]
  
# Create a dataframe using the two lists
df_days_calories = pd.DataFrame(
    { 'days' : days , 
     'casual': casual , 
     'members': members})
  
ax = plt.gca()
  
#use plot() method on the dataframe
df_days_calories.plot( x = 'days' , y = 'casual', ax = ax )
df_days_calories.plot( x = 'days' , y = 'members' , ax = ax )
```
